```python
# Load the saved model from the file
with open('/content/drive/MyDrive/JIS_2.0/xgb_model.pkl', 'rb') as file:
    loaded_model = pickle.load(file)
```

```python
import joblib
import pandas as pd

# Load the validation dataset
validation_data = pd.read_csv(r'/content/drive/MyDrive/JIS_2.0/output.csv')

# Preprocess the validation data
X_validation = validation_data.drop("Class", axis=1)
y_validation = validation_data["Class"]

# Make predictions on the validation data using each model
XG_Boost = loaded_model.predict(X_validation)

# Combine the predictions into a DataFrame
validation_results = pd.DataFrame({
    "XG Boost": XG_Boost,
    "Actual_L": y_validation
})

# Save the validation results to a CSV file
validation_results.to_csv("validation_results.csv", index=False)
```
