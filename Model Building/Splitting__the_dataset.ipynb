```python
df = pd.read_csv('/content/drive/MyDrive/JIS_2.0/CreditCard.csv')
```

```python
df.sample(5)
```

```python
df_fraud = df[df.Class == 1]
```

```python
df_NotFraud = df[df.Class == 0]
```

```python
# Concatenate the data frames vertically
merged_df = pd.concat([df_fraud[:10], df_NotFraud[:20]], ignore_index=True)
merged_df = pd.DataFrame(merged_df)
```

```python
merged_df
```

```python
Final_df = merged_df.sample(30)
```

```python
Final_df.to_csv('output.csv', index=False)
```

```python
# Merge the dataframes
merged_df1 = df.merge(merged_df, how='left', indicator=True)

# Filter out the rows where the entire row matches
filtered_df = merged_df1[merged_df1['_merge'] != 'both']

# Drop the '_merge' column if needed
filtered_df = filtered_df.drop('_merge', axis=1)
```

```python
filtered_df.to_csv('training.csv', index=False)
```
